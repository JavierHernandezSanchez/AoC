## Parte 1

en el mapa se pueden ver espacios vacíos ".", espejos "\" y "/" y splitters "|" y "-". Un rayo de luz entra por la esquina superior izquierda, en los espacioes en blanco sigue sin cambiar su dirección, en los espejos se refleja 90 grados y en los splitters se generan dos rayos en las direcciones del splitter.

una celda está energizada si pasa un rayo por ella. Cuantas celdas hay energizadas=?

```python
import time
from datetime import timedelta
```

```python
def get_input(filename):
    mirror_map = []
    with open(filename) as file:
        for line in file:
            curr = [c for c in line.strip()]
            mirror_map.append(curr)
    return mirror_map
```

```python
def print_grid(grid):
    for i in range(len(grid)):
        for j in range(len(grid[0])):
            print(grid[i][j], end= ' ')
        print()
    print()
```

```python
def create_energized_map(mirror_map):
    energized = []
    covered = []
    
    for i in range(len(mirror_map)):
        energized.append(['.' for _ in range(len(mirror_map[0]))])
        
    beam = [(0, 0), (0, 1)]
    energized[0][0] = '#'
    pila = [beam]
    
    h = len(mirror_map)
    w = len(mirror_map[0])
    
    while len(pila) > 0:
        beam = pila.pop()
        if beam in covered:
            continue
        else:
            covered.append(beam)
            
        f, c = beam[0]
        df, dc = beam[1]
        
        nf = f + df
        nc = c + dc
        
        if nf >= 0 and nf < h and nc >= 0 and nc < w:
            energized[nf][nc] = '#'
            
            if mirror_map[nf][nc] == '.':
                pila.append([(nf, nc), (df, dc)])
            
            elif mirror_map[nf][nc] == '|' and dc != 0:
                pila.append([(nf, nc), (1, 0)])
                pila.append([(nf, nc), (-1, 0)])            
            elif mirror_map[nf][nc] == '|' and dc == 0:
                pila.append([(nf, nc), (df, dc)])
                
            elif mirror_map[nf][nc] == '-' and df != 0:
                pila.append([(nf, nc), (0, 1)])
                pila.append([(nf, nc), (0, -1)])                
            elif mirror_map[nf][nc] == '-' and df == 0:
                pila.append([(nf, nc), (df, dc)])
                
            elif mirror_map[nf][nc] == '\\' and df == 1:
                pila.append([(nf, nc), (0, 1)])                
            elif mirror_map[nf][nc] == '\\' and df == -1:
                pila.append([(nf, nc), (0, -1)])                
            elif mirror_map[nf][nc] == '\\' and dc == 1:
                pila.append([(nf, nc), (1, 0)])                
            elif mirror_map[nf][nc] == '\\' and dc == -1:
                pila.append([(nf, nc), (-1, 0)]) 
                
            elif mirror_map[nf][nc] == '/' and df == 1:
                pila.append([(nf, nc), (0, -1)])                 
            elif mirror_map[nf][nc] == '/' and df == -1:
                pila.append([(nf, nc), (0, 1)])                 
            elif mirror_map[nf][nc] == '/' and dc == 1:
                pila.append([(nf, nc), (-1, 0)])                 
            elif mirror_map[nf][nc] == '/' and dc == -1:
                pila.append([(nf, nc), (1, 0)])
    return energized
```

```python
def count_energy(energized_map):
    count = 0
    for i in energized_map:
        for j in i:
            if j == '#':
                count += 1
    return count
```

```python
mirror_map = get_input('test.txt')
energized = create_energized_map(mirror_map)
count = count_energy(energized)
count
```

```python
mirror_map = get_input('input.txt')
energized = create_energized_map(mirror_map)
count = count_energy(energized)
count
```

## Parte 2

el rayo puede empezar en cualquiera de los bordes del mapa apuntando hacia el lado contrario. Encontrar la configuración que maximiza la cantidad de celdas energizadas.

```python
def create_energized_map2(mirror_map, beam):
    energized = []
    covered = []
    
    for i in range(len(mirror_map)):
        energized.append(['.' for _ in range(len(mirror_map[0]))])
        
    energized[beam[0][0]][beam[0][1]] = '#'
    pila = [beam]
    
    h = len(mirror_map)
    w = len(mirror_map[0])
    
    while len(pila) > 0:
        beam = pila.pop()
        if beam in covered:
            continue
        else:
            covered.append(beam)
            
        f, c = beam[0]
        df, dc = beam[1]
        
        nf = f + df
        nc = c + dc
        
        if nf >= 0 and nf < h and nc >= 0 and nc < w:
            energized[nf][nc] = '#'
            
            if mirror_map[nf][nc] == '.':
                pila.append([(nf, nc), (df, dc)])
            
            elif mirror_map[nf][nc] == '|' and dc != 0:
                pila.append([(nf, nc), (1, 0)])
                pila.append([(nf, nc), (-1, 0)])            
            elif mirror_map[nf][nc] == '|' and dc == 0:
                pila.append([(nf, nc), (df, dc)])
                
            elif mirror_map[nf][nc] == '-' and df != 0:
                pila.append([(nf, nc), (0, 1)])
                pila.append([(nf, nc), (0, -1)])                
            elif mirror_map[nf][nc] == '-' and df == 0:
                pila.append([(nf, nc), (df, dc)])
                
            elif mirror_map[nf][nc] == '\\' and df == 1:
                pila.append([(nf, nc), (0, 1)])                
            elif mirror_map[nf][nc] == '\\' and df == -1:
                pila.append([(nf, nc), (0, -1)])                
            elif mirror_map[nf][nc] == '\\' and dc == 1:
                pila.append([(nf, nc), (1, 0)])                
            elif mirror_map[nf][nc] == '\\' and dc == -1:
                pila.append([(nf, nc), (-1, 0)]) 
                
            elif mirror_map[nf][nc] == '/' and df == 1:
                pila.append([(nf, nc), (0, -1)])                 
            elif mirror_map[nf][nc] == '/' and df == -1:
                pila.append([(nf, nc), (0, 1)])                 
            elif mirror_map[nf][nc] == '/' and dc == 1:
                pila.append([(nf, nc), (-1, 0)])                 
            elif mirror_map[nf][nc] == '/' and dc == -1:
                pila.append([(nf, nc), (1, 0)])
    return energized
```

```python
def try_all_directions(mirror_map):
    start = time.time()
    
    h = len(mirror_map)
    w = len(mirror_map[0])
    max_energized = 0
    
    for i in range(w):
        energized = create_energized_map2(mirror_map, [(0, i), (1, 0)])
        count = count_energy(energized)
        if count > max_energized:
            max_energized = count
            
        energized = create_energized_map2(mirror_map, [(h-1, i), (-1, 0)])
        count = count_energy(energized)
        if count > max_energized:
            max_energized = count
            
        elapsed = time.time() - start
        print(f'horizontal {i} from {w}; max energized: {max_energized} ({str(timedelta(seconds=elapsed))})')
            
    for i in range(h):
        energized = create_energized_map2(mirror_map, [(i, 0), (0, 1)])
        count = count_energy(energized)
        if count > max_energized:
            max_energized = count
            
        energized = create_energized_map2(mirror_map, [(i, w - 1), (0, -1)])
        count = count_energy(energized)
        if count > max_energized:
            max_energized = count
        
        elapsed = time.time() - start
        print(f'vertical {i} from {h}; max energized: {max_energized} ({str(timedelta(seconds=elapsed))})')
    return max_energized
```

```python
mirror_map = get_input('test.txt')
max_energized = try_all_directions(mirror_map)
max_energized
```

```python
mirror_map = get_input('input.txt')
max_energized = try_all_directions(mirror_map)
max_energized
```
