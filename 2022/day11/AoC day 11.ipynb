Each monkey has several attributes:

- Starting items lists your worry level for each item the monkey is currently holding in the order they will be inspected.
- Operation shows how your worry level changes as that monkey inspects an item. (An operation like new = old * 5 means that your worry level after the monkey inspected the item is five times whatever your worry level was before inspection.)
- Test shows how the monkey uses your worry level to decide where to throw an item next.
    - If true shows what happens with an item if the Test was true.
    - If false shows what happens with an item if the Test was false.

After each monkey inspects an item but before it tests your worry level, your relief that the monkey's inspection didn't damage the item causes your worry level to be divided by three and rounded down to the nearest integer.

The monkeys take turns inspecting and throwing items. On a single monkey's turn, it inspects and throws all of the items it is holding one at a time and in the order listed. Monkey 0 goes first, then monkey 1, and so on until each monkey has had one turn. The process of each monkey taking a single turn is called a round.

When a monkey throws an item to another monkey, the item goes on the end of the recipient monkey's list. A monkey that starts a round with no items could end up inspecting and throwing many items by the time its turn comes around. If a monkey is holding no items at the start of its turn, its turn ends.

```python
class Monkey:
    def __init__(self, monkey_id, items, op, strop, divisor, monkey_true, monkey_false):
        self.monkey_id = monkey_id
        self.items = items
        self.op = op
        self.strop = strop
        self.divisor = divisor
        self.monkey_true = monkey_true
        self.monkey_false = monkey_false
        self.inspected_items = 0
        
    def __str__(self):
        monkey_str = f'Monkey {self.monkey_id}:\n'
        monkey_str += f'Items: {self.items}\n'
        monkey_str += f'Operation: {self.op}\n'
        monkey_str += f'Operation str: {self.strop}\n'
        monkey_str += f'Test: divisible by {self.divisor}\n'
        monkey_str += f'\tIf true: throw to monkey {self.monkey_true}\n'
        monkey_str += f'\tIf false: throw to monkey {self.monkey_false}\n'
        monkey_str += f'\tInspected items {self.inspected_items}\n'
        
        return monkey_str
    
    def __repr__(self):
        return self.__str__()
    
    def operations(self):
        self.items = [self.op(i) for i in self.items]
        #old = self.items[0]
        #new = self.op(old)

        #self.items[0] = new
        self.inspected_items += len(self.items)
        
    def update_worry_levels(self):
        self.items = [x // 3 for x in self.items]        
        
    def get_destinations(self):
        dest = [self.monkey_true if x % self.divisor == 0 else self.monkey_false for x in self.items]
        return dest
#         if self.items[0] % self.divisor == 0:
#             dest = self.monkey_true
#         else:
#             dest = self.monkey_false
#         return dest
        
    def receive_item(self, item):
        self.items.append(item)
        
    def get_item(self):
        return self.items.pop(0)
```

```python
def read_input(file_name):
    monkey_id = '*'
    items = []
    op = None
    divisor = 0
    monkey_true = -1
    monkey_false = -1
    
    monkeys = []
    
    with open(file_name) as file:
        for line in file:
            text = line.strip()
            if text == '':
                continue
                
            if text.startswith('Monkey'):
                _, idstr = text.split(' ')
                monkey_id = idstr[:-1]                                
            elif text.startswith('Starting'):
                _, itemstr = text.split(':')
                for i in itemstr.split(','):
                    items.append(int(i.strip()))
            elif text.startswith('Operation'):
                _, opstr = text.split(':')
                _, opstr = opstr.split('=')
                terms = opstr.split()
                                
                if terms[1] == '+' and terms[2] == 'old':                    
                    op = eval('lambda x: x + x')
                    strop = 'lambda x: x + x'
                elif terms[1] == '+' and terms[2] != 'old':
                    op = eval(f'lambda x: x + {int(terms[2])}')
                    strop = f'lambda x: x + {int(terms[2])}'
                elif terms[1] == '*' and terms[2] == 'old':
                    op = eval('lambda x: x * x')
                    strop = 'lambda x: x * x'
                elif terms[1] == '*' and terms[2] != 'old':
                    op = eval(f'lambda x: x * {int(terms[2])}')
                    strop = f'lambda x: x * {int(terms[2])}'
                else:
                    print('WTF')
                    
            elif text.startswith('Test'):
                _, divstr = text.split('by')
                divisor = int(divstr)
            elif text.startswith('If true'):
                _, monkeystr = text.split('monkey')
                monkey_true = int(monkeystr)
            elif text.startswith('If false'):
                _, monkeystr = text.split('monkey')
                monkey_false = int(monkeystr)
                new_monkey = Monkey(monkey_id, items, op, strop, divisor, monkey_true, monkey_false)
                monkeys.append(new_monkey)
                items = []                
            else:                
                print(f'{text} --> WTF!')
    return monkeys
```

```python
from math import gcd
def printv(string, verbose):
    if verbose:
        print(string)
        
def run_round(monkeys, verbose=False):
    for monkey in monkeys:
        monkey.operations()
        monkey.update_worry_levels()
        dest = monkey.get_destinations()
        
        for d in dest:        
            item = monkey.get_item()            
            monkeys[d].receive_item(item)

def get_monkey_business(monkeys, rounds):
    divisors = [m.divisor for m in monkeys]
    lcm = 1
    for i in divisors:
        lcm = lcm*i//gcd(lcm, i)

    for i in range(rounds):
        run_round(monkeys, False)
        #we keep divisibility but we skip using very big numbers
        for m in monkeys:
            m.items = [i % lcm for i in m.items]
    monkey_inspections = [m.inspected_items for m in monkeys]
    sorted_inspections = sorted(monkey_inspections, reverse=True)
    monkey_business = sorted_inspections[0] * sorted_inspections[1]
    return monkey_business
```
